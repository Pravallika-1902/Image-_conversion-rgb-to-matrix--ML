```python
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread(r"C:\Users\Pravallika\OneDrive\Documents\pawan.jpg")
```

```python
# Convert BGR to matrix
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# Display the image
plt.imshow(img_rgb)
plt.title("The Original Image")
plt.axis("off")
plt.show()
```
